```python
import pandas as pd
import numpy as np
import preprocessor as p
import re
import nltk
import string
import json

nltk.download('stopwords')
nltk.download('punkt')
```

```python
# read data excel
ds = pd.read_excel('Dataset_Label_New.xlsx')
ds.to_numpy()
ds
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# setup stopwords
stopwords = set(stopwords.words('indonesian'))
additional_stopwords = ['g', 'yg', 'tdk', 'gak', 'x', 'nya', 'rt', 'no', 'kk', 'sm', 'tp', 'pls', 'cb', 'sgt', 'jg', 'y']
new_stopwords        = stopwords.union(additional_stopwords)
print(stopwords)
# end

# setup stemmer kata
factory = StemmerFactory()
stemmer = factory.create_stemmer()
# end

def caseFolding(text):
    text = str(text).lower()
    text = ' '.join(re.sub("[@#][A-Za-z0-9_]+|([^0-9A-Za-z \t])|(\w+:\/\/\S+)+|<.*?>|^https?:\/\/.*[\r\n]*"," ", text).split(' '))
    text = re.sub(r"\d+", "", text)
    return text

def tokenizing(text):
    text = nltk.tokenize.word_tokenize(text)
    return text

def filtering(token):
    filter_ = [t for t in token if t not in new_stopwords]
    return " ".join(filter_)

def stemming(filtered_token):
    stemmed = stemmer.stem(filtered_token)
    return stemmed
```

```python
#Menjalankan fungsi preprocessing
dataset = list()

for index, row in ds.iterrows():
    print('Proccessing Index '+str(index)+'... ', end = '')
    hasil        = row['Text']
    _casefolding = caseFolding(hasil)
    _tokenizing  = tokenizing(_casefolding)
    _filtering   = filtering(_tokenizing)
    _stemming    = stemming(_filtering)
    
        
    print('Done...')
    dataset.append({        
        "Text": row['Text'],
        "Label": row['Label'],
        "Stem_Text": _stemming
    })
```

```python
ds2 = pd.DataFrame(dataset)
ds2.to_excel('Data_Proses_New.xlsx')
ds2.head(10)
```
